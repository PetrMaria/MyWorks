```python
text = input('Введите текст: ') 
long_word = 0 
count = 0 

for symbol in text: 
  if symbol != ' ': 
    count += 1 
  else: 
    if count > long_word: 
      long_word = count 
    count = 0 
if count > long_word: 
  long_word = count 
  
print('Самое длинное слово, букв: ', long_word)
```

```python
milk = 0
for count in range(10):
  place = input('Стойло свободно? (a-да/b-нет): ')
  if place == 'a':
    milk += 2 * (count + 1)
  if place == 'b':
    milk += 0
print('Молока всего: ', milk)
```

```python
word = input('Введите зашифрованое слово: ')
sum1 = ' '
sum2 = ' '
count = 0

for text in word:
  count += 1
  if (count % 2 == 1):
    sum1 += text
  else:
    sum2 = text + sum2

print('Расшифрованое слово', sum1 + sum2)
```

```python
reversed_string = '' 
string = input('Введите строку: ') 
for symbol in string: 
  reversed_string = symbol + reversed_string 
if string == reversed_string: 
    print('Да, это палиндром!') 
else: 
  print('Нет, это не палиндром!')
```

```python
total_debtor = int(input('Введите количество должников: '))
total_sum = 0
for debtor in range(0, total_debtor, 5):
  print('Должник с номером: ', debtor)
  sum = int(input('Сколько должны? '))
  total_sum += sum
print('Общая сумма долга: ', total_sum)
```

```python
reverse_timer = int(input('Введите время разогрева: ')) 
for time in range(reverse_timer, -1, -1): 
    if time == 0: 
        print('Ваша еда готова, осторожно горячo!') 
        break 
    print('Осталось', time, 'секунд') 
    stop = int(input('Вы готовы забрать еду? 1-Да, 0-Нет: ')) 
    if stop == 1: 
        print('Ваша еда готова, можете забрать') 
        print('Таймер был прерван на', time, 'секунде') 
        break 
    else: 
        print('Продолжаем греть')
```

```python
a = int(input('Введите первое число: '))
b = int(input('Введите второе число: '))
c = int(input('Введите третье число: '))
sum = 0
count = 0
for number in range(a, b +1):
  if number % c == 0:
    count += 1
    sum += number
print(sum / count)
```

```python
start = int(input('Введите начало отрезка: '))
stop = int(input('Введите конец отрезка: '))
step = int(input('Введите шаг: '))
for x in range(stop, start - 1, step):
  y = x ** 3 + 2 * x ** 2 - 4 * x + 1
  print('В точке', x, 'функция равна: ', y)
  print()
```

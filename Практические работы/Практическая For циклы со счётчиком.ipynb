```python
number = int(input('Введите число: '))
factorial = 1
for search in range(1, number + 1):
  factorial *= search
print('Факториал числа', number, 'Равен', factorial)
```

```python
students = int(input('Введите число учеников: '))
five = 0
four = 0
three = 0
for i in range (1,students):
  n = int(input('Введите оценку ученика: '))
  if n == 5:
    five +=1
  elif n == 4:
    four += 1
  elif n == 3:
    three += 1
if five > four and five > three:
  print('Отличников сегодня больше')
elif four > five and three:
  print('Хорошистов сегодня больше')
elif three > five and three > four:
  print('Троечников сегодня больше')
```

```python
a = int(input('Введите первое число: '))
b = int(input('Введите второе число: '))
summ = 0
count_mult = 0
for i in range(a, b + 1):
  if i % 3 == 0:
    summ += i
    count_mult += 1
print('Среднее арифмитическое: ', summ / count_mult)
```

```python
count_card = int(input('Введите количество карточек: '))
summ = 0
for i in range(1, count_card + 1):
  summ += i
for i in range(count_card - 1):
  summ -= int(input('Введите номер оставшейся карточки: '))
print('Номер пропавшей карточки: ', summ)
```

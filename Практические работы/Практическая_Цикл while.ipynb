```python
positive = 0
negative = 0
n = 1
while n != 0:
  n = int(input('Введите число: '))
  if n > 0:
    positive += 1
  elif n < 0:
    negative += 1
  if n == 0:
    break
print('Количество положительных оценок: ', positive)
print('Количество отрицательных оценок: ', negative)
```

```python
print('Начался 8-часовой рабочий деь')
hours = 0
task = 0
call = 0
while hours < 8:
  hours += 1
  print(hours, '-й час')
  task += int(input('Сколько задач решит Максим? '))
  call += int(input('Звонит жена. Взять трубку? (1 - да, 0 - нет) '))
  if call == True and hours == 8:
    print('Рабочий день закончился. Всего выполнено задач:', task)
    print('Нужно зайти в магазин.')
    break
  elif call == False and hours == 8:
    print('Рабочий день закончился. Всего выполнено задач: ', task)
    break
```

```python
x = int(input('Введите сумму вклада: '))
p = int(input('Введите процент: '))
y = int(input('Введите желаемую сумму: '))
years = 0
while x < y:
  x = x + (x * p / 100)
  x = int(100 * x) / 100 
  years += 1 
print('Понадобится ', years, 'лет.')
```

```python
father = 8
count = 0
son = int(input('Введите число:  '))
while son != father:
  count += 1
  if son < father:
    print('Число меньше, чем нужно. Попробуйте ещё раз!')
    son = int(input('Введите число:  '))
  else:
    print('Число больше, чем нужно. Попробуйте ещё раз!')
    son = int(input('Введите число:  '))
print('Вы угадали! Количество попыток: ' , count)
```

```python
min = 1
max = 100
count = 1
while True:
  n = (min + max) // 2
  print('Загаданное число равно, меньше или больше?' , n)
  answer = int(input('1 - равно, 2 - меньше, 3 - больше: '))
  if answer == 1:
    print('Вы угадали! Число попыток: ', count)
    break
  elif answer == 2:
    max = n
  elif answer == 3:
    min = n
  count += 1
```
